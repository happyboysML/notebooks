```julia
using Plots
using LinearAlgebra
```

```julia
Gauss(x, μ, σ) = 1/(sqrt(2π)*σ) * exp.(-(x .- μ).^2/(2σ^2)) 

std_gauss = x -> Gauss(x, 0, 1)

x = [i for i in -4:0.1:4]

plot(x, std_gauss(x))
```

# Multivariate Gaussian
$$
\mathcal{N}(\bf{x}|\mu, \Lambda) = \frac{1}{(2 \pi)^{D/2}}|\Lambda|^{1/2} exp \left\{-\frac{1}{2} (\bf{x} - \mu)^T \Lambda (\bf{x} - \mu) \right\}
$$

```julia
MultiGauss(x, μ, Λ) = 1/(2π)^(size(x, 2)/2) * det(Λ)^0.5 * exp(-0.5(x .- μ)' * Λ * (x .- μ))
std_gauss2 = (x1, x2) -> MultiGauss([x1, x2], [0, 0], I)

contour(x, x, std_gauss2,
    aspect_ratio=1,
    c="blue"
)
```

# The von Mises distribution
The blue contours indicate 2D Gaussian.
The red ring is a unit circle.
The right figure shows the normalized 2D Gaussian values on the red circle.
It's a distribution called the von Mises distribution.
It goes steeper as far from 2D Gaussian's center.

$$
p(\theta|\theta_0, m) = \frac{1}{2\pi I_0 (m)} exp\{ m cos(\theta - \theta_0)\}
$$

```julia
μ_gauss2(μ) = (x1, x2) -> MultiGauss([x1, x2], μ, I)

anim = Animation()
θ = [i for i in 0:0.01:2π]
θ_0 = [i for i in 0:0.05:2π]
for i in 1:size(θ_0, 1)
    r = 3
    μ = [r*cos(θ_0[i]), r*sin(θ_0[i])]
    
    # von Mises distribution
    d = μ_gauss2(μ).(cos.(θ), sin.(θ))
    vonMises = d/sum(d)
    
    # following codes are just for plot
    l = @layout [
        a b 
    ]
    contour(x, x, μ_gauss2(μ),
        aspect_ratio=1,
        c="blue",
        label="2D Gauss",
        xlabel="x1",
        ylabel="x2",
        xlim=[-4, 4],
        ylim=[-4, 4],
        colorbar=false,
    )
    p1 = plot!(cos.(θ), sin.(θ),
        aspect_ratio=1,
        c="red",
        label="unit circle",
    )
    p2 = plot(θ, vonMises,
        xlabel="θ",
        ylabel="p(θ)",
        ylim=[0, 0.009],
    )
    frame(
        anim,
        plot(p1, p2,
            layout = l,
            legend = :bottomleft
        )
    )
end
gif(anim, "vonMises.gif", fps=15)
```
